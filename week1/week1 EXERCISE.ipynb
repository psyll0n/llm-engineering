To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

````python
# imports
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
from ollama import Client as ollamaClient
````

````python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
````

````python
# set up environment
load_dotenv()
openai_client = OpenAI()  # This will automatically use OPENAI_API_KEY from environment
ollama_client = ollamaClient(host='http://localhost:11434')
````

````python
# Systen Prompts
system_prompt = "You are a helpful technical tutor who answers questions about python code, software engineering, data science and LLMs"
````

````python
# Get gpt-4o-mini to answer, with streaming
def get_gpt_answer(question):
    try:
        messages = [
            {
                "role": "system",
                "content": "You are a helpful technical expert who explains code clearly and concisely."
            },
            {
                "role": "user",
                "content": question
            }
        ]

        stream = openai_client.chat.completions.create(
            model=MODEL_GPT, 
            messages=messages,
            stream=True
        )
        
        response = ""
        display_handle = display(Markdown(""), display_id=True)
        for chunk in stream:
            if chunk.choices[0].delta.content:
                response += chunk.choices[0].delta.content
                response = response.replace("```","").replace("markdown", "")
                update_display(Markdown(response), display_id=display_handle.display_id)
        
        return response
    except Exception as e:
        return f"Error getting GPT response: {str(e)}"

def get_llama_answer(question):
    try:
        response = ollama_client.chat(
            model=MODEL_LLAMA,
            messages=[
                {
                    "role": "system",
                    "content": "You are a helpful technical expert who explains code clearly and concisely."
                },
                {
                    "role": "user",
                    "content": question
                }
            ]
        )
        answer = response['message']['content']
        display(Markdown(answer))
        return answer
    except Exception as e:
        return f"Error getting Llama response: {str(e)}"

def ask_question():
    """Get user input and create formatted prompts"""
    question = input("Please enter your technical question: ")
    
    messages = {
        'gpt': [
            {
                "role": "system",
                "content": "You are a helpful technical expert who explains code clearly and concisely."
            },
            {
                "role": "user",
                "content": f"Please explain: {question}"
            }
        ],
        'llama': [
            {
                "role": "system",
                "content": "You are a helpful technical expert who explains code clearly and concisely."
            },
            {
                "role": "user",
                "content": f"Please explain: {question}"
            }
        ]
    }
    return question, messages

def get_answers():
    """Get answers from both models"""
    question, messages = ask_question()
    
    print("\nGetting GPT response...")
    gpt_answer = get_gpt_answer(question)
    
    print("\nGetting Llama response...")
    llama_answer = get_llama_answer(question)
    
    return gpt_answer, llama_answer

# Interactive usage
if __name__ == "__main__":
    while True:
        gpt_response, llama_response = get_answers()
        
        continue_asking = input("\nWould you like to ask another question? (y/n): ")
        if continue_asking.lower() != 'y':
            break
````
